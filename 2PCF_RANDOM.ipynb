```python
import numpy as np
import matplotlib.pyplot as plt
import math
from time import time
Ti=time()
```

```python
x = np.random.rand(5000)*100
y = np.random.rand(5000)*100
x2 = np.random.rand(5000)*100
y2 = np.random.rand(5000)*100
```

```python
d_random = [] #List of distance for random data.

for i in range(len(x)):
    for j in range(i+1,len(x)):
        d_random_ij=np.sqrt((x[i]-x[j])**2+(y[i]-y[j])**2)
        d_random.append(d_random_ij)

d_random2 = [] #List of distance for random data.

for i in range(len(x2)):
    for j in range(i+1,len(x2)):
        d_random2_ij=np.sqrt((x2[i]-x2[j])**2+(y2[i]-y2[j])**2)
        d_random2.append(d_random2_ij)
        
# We create the histogram for random data.
RR = plt.hist(d_random, bins=140,  facecolor='royalblue', label=r'Distances', linewidth=0, alpha=0.5)[0]
RR2 = plt.hist(d_random2, bins=140,  facecolor='royalblue', label=r'Distances', linewidth=0, alpha=0.5)[0]
plt.xlabel('Distance')
plt.ylabel('Number of points')
plt.legend(loc=1)
plt.title('Distances of points in random data')
plt.show()

Tf=time()
print Tf-Ti
```

```python
DP = (RR/(RR2))-1 
max_dist = 100*np.sqrt(2)

plt.plot(DP,color='r',label=r'$DataRandoms$')
plt.axhline(y=0,linestyle='--',linewidth=1,color='k')
plt.title('Davis-Peebles Estimator')
plt.xlabel(r'$r$')
plt.ylabel(r'$\epsilon(r)$')
plt.xlim(0,max_dist)
plt.legend()
plt.show()
```
